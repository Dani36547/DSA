#### import the necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Read the dataset to python enviornment

```python
data = pd.read_excel(r"F:\data sets\iris.xls")
```

#### Display the first five rows of the dataset

```python
data.head()
```

#### Summary of the dataset

```python
data.info()
```

#### Display the columns in the dataset

```python
data.columns
```

#### Mean value of numerical columns in the dataset

```python
print("mean value of SL =",data['SL'].mean())
print("mean value of SW =",data['SW'].mean())
print("mean value of PL =",data['PL'].mean())
print("mean value of PW =",data['PW'].mean())
```

#### Check for null values

```python
data.isnull().sum()
```

### Data Visualizations

```python
# Countplot for classification

sns.countplot(x=data['Classification'])
plt.show()
```

#### Relationship between variables

```python
# Comparing Sl and SW
sns.scatterplot(x='SL',y='SW', hue='Classification', data=data)
plt.show()
```

```python
# Comparing PL and PW
sns.scatterplot(x='PL',y='PW', hue='Classification',data=data)
plt.show()
```

#### Plot all the columns relationships using a pairplot

```python
sns.pairplot(data)
plt.show()
```

##### Frequency distributions of data 

```python
plt.figure(figsize=(8,6))

plt.subplot(2,2,1)
sns.boxplot(x='Classification', y='SL', data=data)

plt.subplot(2,2,2)
sns.boxplot(x='Classification', y='SW', data=data)

plt.subplot(2,2,3)
sns.boxplot(x='Classification', y='PL', data=data)

plt.subplot(2,2,4)
sns.boxplot(x='Classification', y='PW', data=data)

plt.tight_layout()
```

#### Checking correlation

```python
sns.heatmap(data.corr(), annot=True)
plt.show()
```
